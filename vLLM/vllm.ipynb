```python
%pip install vllm

from vllm import LLM
from vllm.utils import CompletionRequest

def main():
    # Tạo đối tượng LLM load mô hình (ví dụ LLaMA-7B)
    llm = LLM(model="decapoda-research/llama-7b-hf")

    # Tạo yêu cầu sinh văn bản
    request = CompletionRequest(prompt="Hello, vLLM! What is AI?", max_tokens=50)

    # Thực hiện sinh văn bản
    response = llm.generate([request])

    # In kết quả trả về
    for batch in response:
        for doc in batch:
            print(doc.text)

if __name__ == "__main__":
    main()
```
