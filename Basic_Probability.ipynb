8. Định bý Bayes
P(A|B) = [P(B|A) * P(A)] / P(B)

Ví dụ:
Túi 1 có 75 đỏ 25 xanh
Túi 2 có 60 đỏ 40 xanh
Túi 3 có 45 đỏ 55 xanh
Xác suất chọn được bi đỏ từ túi thứ 1 là bao nhiêu?
P(B1) = 1/3 = P(B2) = P(B3)
P(R/B1) = 75,
P(R/B2) = 60,
P(R/B3) = 45,
p(R) = P(R/B1)/P(B1) + P(R/B1)/P(B1) + P(R/B1)/P(B1) = 0.6

P(B1/R) = P(R/B1)P(B1)/P(R) = 5/12


Ừ, vui là cậu hỏi đấy! Bắt đầu với trực giác đằng sau bài chứng minh nhé: nó đơn giản chỉ là đổi mẫu số trong một phân số thôi.

Định lý Bayes nói rằng P(A|B) = P(B|A)P(A)/P(B). Mở rộng P(B|A) ở vế phải ra thì cậu sẽ được:

P(A|B) = [P(AB)/P(A)][P(A)/P(B)]

Rút gọn P(A) đi thì cậu sẽ được P(AB)/P(B), tức là P(A|B). Vì vậy định lý Bayes "nói" rằng P(A|B) và P(B|A) có cùng tử số P(AB) và mẫu số ngược nhau (P(B) so với P(A)).

Cậu có thể hiểu trực giác hình học hơn một chút bằng cách vẽ vài hình. Một ví dụ hay là video này . Nó không giúp ích nhiều hơn (không giống như, ví dụ, trực giác hình học đằng sau khái niệm xác suất có điều kiện mà video giải thích rất tốt). Nhưng cái này cũng giúp ích phần nào.

Bức tranh tổng quan: Cậu đã học được một số dữ liệu B. Khi nào thì điều đó làm cho giả thuyết A của cậu có khả năng xảy ra hơn? À, khi B làm cho A có khả năng xảy ra hơn! Điều đó gợi ý:

(Mối quan hệ tuyến tính giữa xác suất hậu nghiệm và khả năng xảy ra): P(A|B) nên tăng tuyến tính với P(B|A).

Cách tự nhiên nhất để làm được điều này là đặt P(A|B) = P(B|A), nhưng điều đó không hoàn toàn đúng. Một mặt, nếu A đã khá có khả năng xảy ra thì xác suất trước đó của A vẫn nên được tính để ủng hộ A. Điều đó gợi ý:

(Mối quan hệ tuyến tính giữa xác suất hậu nghiệm và xác suất trước của giả thuyết): P(A|B) nên tăng tuyến tính với P(A).

Cách tự nhiên nhất để thỏa mãn cả hai ràng buộc là đặt P(A|B) = P(B|A)P(A). Nhưng điều đó cũng không hoàn toàn đúng. Giả sử B đã khá có khả năng xảy ra. Thì việc học được B không nên làm tăng xác suất của A nhiều, bởi vì ở chừng mực mà chúng ta biết rằng B đã khá có khả năng xảy ra, xác suất trước P(A) của chúng ta đã tính đến hầu hết tác động của B lên A rồi. Điều này gợi ý:

(Mối quan hệ nghịch biến giữa xác suất hậu nghiệm (của giả thuyết) và xác suất trước của dữ liệu): P(A|B) thay đổi nghịch với P(B).

Kết hợp tất cả lại thì ta được P(A|B) = P(B|A)P(A)/P(B). Và điều đó hóa ra là một định lý.

Cái này có giúp ích gì không? Đôi khi khó biết cách nào sẽ là cách thuyết phục để giải thích trực giác đằng sau một định lý, vì vậy cho mình biết nếu cậu đang tìm kiếm các ứng dụng hoặc các dạng khác của định lý Bayes hoặc có câu hỏi cụ thể hay gì đó tương tự nhé.
