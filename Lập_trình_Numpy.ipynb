```python
# Ứng dụng trong softmax function
x = np.array([1,2,3])
print(x)
m = max(x)
print(m)

X_m = x - m
print(X_m)

exp_x = np.exp(X_m)
print(exp_x)

sum_exp_x = np.sum(exp_x)
print(sum_exp_x)

soft_max = exp_x/sum_exp_x
print(soft_max)
```

```python
import cv2

#Read the image
image = cv2.imread('/content/image2.png', 0)

print(type(image)) #ndarray

print(image)

plt.imshow(image, cmap = 'gray')

image = image.astype(float)

image = image + 50
image = np.where(image<0, 0, image)
image = np.where(image>255, 255, image)

image = image.astype(np.uint8)
plt.imshow(image, cmap = 'gray')
```
