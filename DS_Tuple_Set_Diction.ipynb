Review: From 1D to 2D List
Lập trình truyền vào hàng mấy cột mấy để lấy bạn ý ra.

Screenshot 2025-06-19 at 20.36.26.png

```python
student_list = ["Vinh", "An", "Toan", "Lam", "Tam", "Chuyen"]
row_class = 3
col_class = 2

def find_student(student_list, r, c):
  index = (r -1)*col_class + (c-1)
  return student_list[index]

print(find_student(student_list, 3, 1))

print(find_student(student_list, 3, 2))
```

Không dùng kiểu dữ liệu 2D thì khá là phức tạp để xử lí bài toán có chứa hàng chứa cột.

```python
row1_student = ["Vinh", "An"]
row2_student = ["Toan", "Lam"]
row3_student = ["Tam", "Chuyen"]

student_list = [row1_student, row2_student, row3_student] #2D List

print(student_list)

def find_student2(student, r, c):
  return student[r-1][c-1]

print(find_student2(student_list, 3, 1)) #Tam
print(find_student2(student_list, 3, 2)) #Chuyen

#Don gian hon
```

```python
#Dictionary: có dạng "key": value
#key không thay đổi được và không được trùng nhau, nhưng value thì thay đổi được

number_dict = {"One":11, "Two":22, "Three": 33, "Four":44}

print(number_dict["One"])#11


a_dict = {str(i):i for i in range(5)}
print(a_dict)
```

```python
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)

#Tạo key và value theo 2 cặp bằng zip
a_dict = dict(zip(tuple1, tuple2))
print(a_dict)
```

```python
d1 = {'a' : [1,2], 'b':5}
d2 = d1.copy()

#d2 thay doi cung se thay doi d1 cho tham chieu cung bo nho
d2['a'][1] = 5

print(d2)
print(d1)

#=> Goi la shallơ copy
```

```python
import copy
d1 = {'a' : [1,2], 'b':5}
d2 = copy.deepcopy(d1)

#d2 khong anh huong d1
d2['a'][1] = 5

print(d2)
print(d1)

#=> Goi la deep copy
```

```python
d1 = {'a' : [1,2], 'b':5}
value = d1.get('a') #Lay gia tri tu dic va khong xoa
print(d1)
print(value)

d1 = {'a' : [1,2], 'b':5}
value2 = d1.pop('a') #Lay ra dic và xoá
print(value2)
print(d1)
```

```python
d1 = {'a' : [1,2], 'b':5}
iemt = d1.popitem() #Lay phan tu cuoi và xoá

print(iemt)
print(d1)
```
