```python
from __future__ import division, print_function, unicode_literals
import numpy as np 
import matplotlib.pyplot as plt

# Fix the shape of y array by transposing it
X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T
y = np.array([[49, 50, 51, 54, 58, 59, 60, 62, 63, 64, 66, 67, 68]]).T  # Remove extra 68 and transpose

# Visualize the data
plt.plot(X, y, 'ro')
plt.axis([140, 190, 45, 75])
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight')
plt.grid()
plt.show()

```

```python
one = np.ones((X.shape[0], 1))  # Create a column of ones for the intercept
Xbar = np.concatenate((one, X), axis = 1)  # Add the column of ones to the feature matrix

# Calculate the weights of the fitting line
A = np.dot(Xbar.T, Xbar)
b = np.dot(Xbar.T, y)
w = np.dot(np.linalg.pinv(A), b)  # Use np.linalg.pinv for pseudo-inverse
print('w:', w)


# Preparing the fitting line
w_0 = w[0, 0]
w_1 = w[1, 0]
x_fit = np.linspace(140, 185, 2)
y_fit = w_0 + w_1 * x_fit

# Visualize the fitting line
plt.plot(X, y, 'ro')
plt.plot(x_fit, y_fit, 'b-')
plt.axis([140, 190, 45, 75])
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight with Fitting Line')
plt.grid()
plt.show()
```

```python
y1 = w_1*155 + w_0
y2 = w_1*160 + w_0
print('Predicted weight for height 155 cm:', y1, 'real number: 52')
print('Predicted weight for height 160 cm:', y2, 'real number: 56')
```
